```python
# -*- coding: utf-8 -*-
"""Jueguito1_v1_0_1.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1KGOU5HKYLDnijO6MAFZ-w7EJeb7FLYQO

# Prueba de juego Nº1: Ta-te-tí

V 1.1.0

Voy a a tratar de hacer una GUI
"""

#instalo librería que sirve para tener interfáz gráfica básica
#!pip install ipywidgets

#importo librerías

import ipywidgets as widgets
from IPython.display import display
import random

# Funciones del juego de Ta-Te-Ti

def crear_tablero():
  tablero = [[" ", " ", " "],
             [" ", " ", " "],
             [" ", " ", " "]]
  return tablero

def mostrar_tablero(tablero):
  for fila in tablero:
    print("| " + " | ".join(fila) + " |")

def es_casilla_vacia(tablero, fila, columna):
  return tablero[fila][columna] == " "

def marcar_casilla(tablero, marca, fila, columna):
  if es_casilla_vacia(tablero, fila, columna):
    tablero[fila][columna] = marca
    return True
  else:
    return False

def hay_ganador(tablero, marca):
  # Verificar filas
  for fila in tablero:
    if fila[0] == fila[1] == fila[2] == marca:
      return True
  # Verificar columnas
  for i in range(3):
    if tablero[0][i] == tablero[1][i] == tablero[2][i] == marca:
      return True
  # Verificar diagonales
  if tablero[0][0] == tablero[1][1] == tablero[2][2] == marca:
    return True
  if tablero[0][2] == tablero[1][1] == tablero[2][0] == marca:
    return True
  return False

def esta_empatado(tablero):
  for fila in tablero:
    for casilla in fila:
      if casilla == " ":
        return False
  return True

def cambiar_turno(turno):
  if turno == "X":
    return "O"
  else:
    return "X"

# Creación del tablero de juego

def crear_game_board():
  # Crea un diseño de cuadrícula para representar el tablero
  board_grid = widgets.GridspecLayout(3, 3)

  # Crea botones individuales para cada casilla del tablero
  buttons = []
  for row in range(3):
    for col in range(3):
      button = widgets.Button(description='', style={'width': '50px', 'height': '50px'})
      button.row = row
      button.col = col
      buttons.append(button)

  # Agrega los botones al diseño de cuadrícula
  for i, button in enumerate(buttons):
    board_grid[button.row, button.col] = button

  return board_grid, buttons # Return the buttons list as well

game_board_widget, buttons = crear_game_board() # Unpack both return values

# Manejo de los movimientos del jugador

def handle_player_move(button):
  global current_player # Tell the function to use the global 'current_player'
  # Obtiene la fila y la columna del botón presionado
  row = button.row
  col = button.col

  # Verifica si la casilla está vacía
  if game_board[row][col] == ' ':
    # Marca la casilla con el símbolo del jugador actual
    game_board[row][col] = current_player

    # Actualiza el texto del botón
    button.description = current_player

    # Verifica si hay un ganador o un empate
    if hay_ganador(game_board, current_player):
      display(widgets.Label(value=f'{current_player} gana!'))
      reset_game()
    elif esta_empatado(game_board):
      display(widgets.Label(value='¡Empate!'))
      reset_game()

    # Cambia de turno
    current_player = 'X' if current_player == 'O' else 'O'

# Conecta los clics de los botones a la función handle_player_move
for button in buttons:
  button.on_click(handle_player_move)

# Función para reiniciar el juego

def reset_game():
  # Limpia el tablero de juego
  for i in range(3):
    for j in range(3):
      game_board[i][j] = ' '

  # Restablece el texto de los botones
  for button in buttons:
    button.description = ''

  # Restablece el jugador actual
  global current_player
  current_player = 'X'

  # Muestra el tablero de juego actualizado
  display(game_board_widget)

# Inicialización y visualización del juego

# Inicializa el tablero de juego
game_board = crear_tablero()

# Establece el jugador inicial
current_player = 'X'

# Muestra el tablero de juego
mostrar_tablero(game_board)

# Muestra el tablero de juego en la interfaz de usuario
display(game_board_widget)
```
